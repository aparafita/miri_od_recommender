```python
import time
from random import random

import selenium
import selenium.webdriver

from selenium.common.exceptions import NoSuchElementException
```

```python
ff = selenium.webdriver.Chrome('/Users/alvaro.parafita/chromedriver')
ff.set_page_load_timeout(15)
```

# Scraping

```python
base_url = 'https://www.tripadvisor.com/Restaurants-g187497-Barcelona_Catalonia.html'
ff.get(base_url)
```

```python
restaurants = {}
```

```python
while True:
    t1 = time.time()
    
    for item in ff.find_elements_by_xpath(r'//div[contains(@class, "listing")]'):
        idx = item.get_attribute('id')
        item = item.find_element_by_xpath(r'./div[@class="shortSellDetails"]/h3[@class="title"]/a')

        restaurants[idx] = { 
            'href': item.get_attribute('href'),
            'name': item.text.strip()
        }

    try:
        next_button = ff.find_element_by_xpath(r'//a[contains(@class, "nav next")]')
        next_button.click()
        
        time.sleep(random() * 2 + 1) # delay between 1 and 3 seconds
    except NoSuchElementException:
        break
        
    t2 = time.time()
    print(t2 - t1)
```

```python
len(restaurants)
```

```python
import json

with open('restaurants.json', 'w') as f:
    f.write(json.dumps(restaurants, indent=2))
```
